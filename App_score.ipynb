```python
joined['Applied at'] = pd.to_datetime(joined['Applied at'], format='%m.%d.%Y %H:%M:%S')

def calculate_score(row):
    if pd.isnull(row['Amount']) or row['External Rating'] == 0:
        return 0
    
    score = 0
    
    # Kryterium: wiek kandydata
    score += ((row['Age'] >= 35) & (row['Age'] <= 55)) * 20
    
    # Kryterium: aplikacja zaÅ‚oÅ¼ona w tygodniu, poza weekendem
    score += ((row['Applied at'].dayofweek) < 5) * 20  
    
    # Kryterium: kandydat w zwiÄ…zku maÅ‚Å¼eÅ„skim
    score += (row['Marital status'] == 'Married') * 20
    
    # Kryterium: kandydat mieszka w Warszawie 
    score += ((row['Location'] == 'Poland, Warsaw') | (row['Location'] == 'Mazowieckie')) * 10
    
    # Kryterium: Score z tabeli industries.csv
    score += row['Score']
    
    # Kryterium: wartoÅ›Ä‡ 'External Rating'
    external_rating = row['External Rating']
    if external_rating >= 7:
        score += 20
    elif external_rating <= 2:
        score -= 20
    
    max_possible_score = 110  
    normalized_score = (score / max_possible_score) * 100

    if normalized_score > 100:
        normalized_score = 100
    
    return normalized_score

joined['New_score'] = joined.apply(calculate_score, axis=1)

# Grupowanie danych wedÅ‚ug tygodnia, w ktÃ³rym zÅ‚oÅ¼ono aplikacjÄ™, i obliczenie Å›redniej oceny zaakceptowanych aplikacji w kaÅ¼dym tygodniu
weekly_avg_score = joined[joined['New_score'] > 0].groupby(joined['Applied at'].dt.isocalendar().week)['New_score'].mean()

# Wizualizacja Å›redniej oceny zaakceptowanych aplikacji 
plt.figure(figsize=(10, 6))
weekly_avg_score.plot(kind='bar', color='skyblue')
plt.title('Åšrednia ocena zaakceptowanych aplikacji w kaÅ¼dym tygodniu')
plt.xlabel('TydzieÅ„')
plt.ylabel('Åšrednia ocena')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
